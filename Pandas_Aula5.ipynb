#Gráficos Continuação


```python
import matplotlib.pyplot as plt #Biblioteca gráfica
import pandas as pd
import numpy as np # Arrays númericos 
import seaborn as sns # biblioteca gráfica extendida da mtplotlib

%matplotlib inline
```

```python
mtcars = pd.read_csv("http://emsapi.esy.es/datasets/mtcars.csv", index_col=0)
```

```python
sns.pairplot(mtcars.loc[:,'mpg':'hp'])
plt.show()
```

# Relações Lineares, Quadráticas e Cúbicas 


```python
x = np.linspace(0, 2, 100) # cria números entre o valor inicial e o final, indicados nos parâmetros
```

```python
# Criando um gráfico Linear
plt.plot(x, x)
```

```python
# Relação Quadrática
plt.plot(x, x**2)
plt.show()
```

```python
# Quadrática completa
x = np.linspace(-2, 2, 100)
```

```python
plt.plot(x, x**2)
plt.show()
```

```python
# Gráficos para relações cúbicas
x = np.linspace(0, 2, 100)
plt.plot(x, x**3)
plt.show()
```

```python
# Gráficos para relações cúbicas compĺetas
x = np.linspace(-2, 2, 100)
plt.plot(x, x**3)
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
plt.plot(x, x**4)
plt.show()
```
